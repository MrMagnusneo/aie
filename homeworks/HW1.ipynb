```python
# Импортируем библиотеку NumPy
import numpy as np
```

```python
# 1. Создание массива из обычного списка
list_data = [1, 2, 3, 4, 5]
array_from_list = np.array(list_data)
print("Массив из списка:", array_from_list)

# 2. Создание массивов с помощью встроенных функций NumPy
zeros_array = np.zeros(5)          # массив из пяти нулей
ones_array = np.ones((2, 3))       # двумерный массив 2x3 из единиц
range_array = np.arange(0, 10, 2)  # массив [0, 2, 4, 6, 8]

print("Массив нулей:", zeros_array)
print("Массив единиц:\n", ones_array)
print("Массив с шагом 2:", range_array)
```

```python
# Выводим форму массива (количество строк и столбцов)
print("Форма массива (shape):", ones_array.shape)

# Выводим тип данных элементов массива
print("Тип данных (dtype):", ones_array.dtype)
```

```python
# Получение отдельного элемента (например, третий элемент, индекс 2)
element = range_array[2]
print("Третий элемент массива:", element)

# Извлечение подмассива с помощью среза (элементы со 2-го по 4-й)
subarray = range_array[1:4]
print("Срез массива:", subarray)
```

```python
# Создадим два массива одинаковой длины
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Поэлементное умножение массивов
product = a * b
print("Поэлементное умножение:", product)

# Вычисление среднего значения одного из массивов
mean_value = a.mean()
print("Среднее значение массива a:", mean_value)
```
